# Challenge 2 - Data Cleaning and Transformation

If your conclusion from the previous challenge is the data need cleaning/transformation, do it in the cells below. However, if your conclusion is the data need not be cleaned or transformed, feel free to skip this challenge. But if you do choose the latter, please provide rationale.

```python
# Your code here
# My conclusions:
# - No categorical data to transform, no missing data to fill up
# - There is some collinearity between values, at least Detergents_Paper and Grocery -> I should drop one column.
# - There are outliners clearly visible -> I should delete the most extreme ones
# - The model is right-skewed -> I should apply some transformation to it
```

```python
# Let's calculate the Variance Inflation Factor to decide which column to delete
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data = pd.DataFrame()
vif_data["feature"] = data.columns
vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(data.shape[1])]
print(vif_data)

# Grocery and Detergents_Paper have the highest VIF. Let's try to delete Grocery, as it has the max value and also many outliners.
data_uncor = data.drop("Grocery", axis=1)

# Let's check the collinearity after dropping the column
plot_correlation_matrix(data_uncor)

# much better, correlation values don't exceed 60%.
```

```python
# Now let's get rid of outliners using the Z-score
from scipy import stats

# calculate zscore and filter dataset
z_scores = np.abs(stats.zscore(data_uncor))
print(z_scores.max())

data_no_outliners = data_uncor[(z_scores < 3).all(axis=1)]

# let's check the result
plot_boxplot(data_no_outliners)

# we got rid of extreme outliners, let's keep some of them that are close to the top whisker, as they might be legitimate data points
```

```python
# Now let's try to unskew the model using a Box-Cox transformation
from scipy import stats

data_unskewed = data_no_outliners.copy()

for column in data_unskewed.columns:
    data_unskewed[column], _ = stats.boxcox(data_unskewed[column] + 1)

plot_boxplot(data_unskewed)

# The whiskers are now generally even on both sides, indicating that the dataset is unskewed
```

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:
from sklearn.preprocessing import StandardScaler

# Your code here:
# let's reload our dataset
data = pd.read_csv("../data/Wholesale customers data.csv")

# Let's scale and train our model using the initial dataset without any preprocessing
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
plot_boxplot(data_scaled)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# UserWarning: KMeans is known to have a memory leak on Windows with MKL, when 
# there are less chunks than available threads. You can avoid it by setting the 
# environment variable OMP_NUM_THREADS=2.
import os
os.environ['OMP_NUM_THREADS'] = "2"

# Your code here:
# inisialise model
k_means = KMeans(init='k-means++', n_clusters=3, n_init=12)

# train model
k_means.fit(data_scaled)

# grab labels and clusters
labels = k_means.predict(data_scaled)
clusters = k_means.labels_.tolist()
data_scaled['Labels'] = clusters
data_scaled
```

```python
# let's calculate the silhoutte score of my KMeans
from sklearn.metrics import silhouette_score

score_kmeans = silhouette_score(data_scaled, data_scaled['Labels'])
print(f"Silhouette Score for KMeans: {score_kmeans}")
```

```python
# let's visualise the elbow score for multiple k values
import os
os.environ['OMP_NUM_THREADS'] = "2"
import warnings
warnings.filterwarnings('ignore')

# Visualise elbow
inertia = []

# Testing 1 to 10 clusters
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)

# plot Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(data_scaled)
labels = kmeans_2.predict(data_scaled)
clusters = kmeans_2.labels_.tolist()
data_scaled['Labels'] = clusters
data_scaled
```

```python
# let's calculate the new silhoutte score of my KMeans using 2 clusters
score_kmeans = silhouette_score(data_scaled, data_scaled['Labels'])
print(f"Silhouette Score for KMeans: {score_kmeans}")

# The silhouette score went down slightly, indicating that a n_clusters of 3
# might give better results
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:
import os
os.environ['OMP_NUM_THREADS'] = "2"
import warnings
warnings.filterwarnings('ignore')

# Visualise elbow
inertia = []

# Testing 1 to 10 clusters
K = range(1, 11)
for k in K:
    print(f"n_clusters = {k}")
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_scaled)
    inertia.append(kmeans.inertia_)
    labels = kmeans.predict(data_scaled)
    data_scaled['Labels'] = kmeans.labels_.tolist()
    score_kmeans = silhouette_score(data_scaled, data_scaled['Labels']) if k > 1 else '/'
    print(f"Silhouette Score: {score_kmeans}")
    plot(data_scaled, 'Detergents_Paper', 'Milk', 'Labels')
    # plot(data_scaled, 'Grocery', 'Fresh', 'Labels')
    # plot(data_scaled, 'Frozen', 'Delicassen', 'Labels')

# plot Elbow curve
plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

**Your comment here**

Based on the graphs, especially the elbow curve, we can see that a k value of 2 or 3 works best for our K-Means model

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Let's perform a grid search over a range of eps and min_samples values and calculate the silhousette score
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

best_score = -1
best_params = {}

# Define parameter ranges
eps_range = np.arange(0.1, 2.0, 0.1)
min_samples_range = range(5, 20)

for eps in eps_range:
    for min_samples in min_samples_range:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(data_scaled)
        
        # Ignore clusters with only noise points
        if len(set(labels)) > 1:
            score = silhouette_score(data_scaled, labels)
            if score > best_score:
                best_score = score
                best_params = {'eps': eps, 'min_samples': min_samples}

print("Best Parameters:", best_params)
print("Best Silhouette Score:", best_score)
```

```python
# let's apply our best parameters
dbscan2 = DBSCAN(eps=1.90, min_samples=5)
dbscan2.fit(data_scaled)
data_scaled['Labels_DBSCAN2'] = dbscan2.labels_
data_scaled['Labels_DBSCAN2'].value_counts()
```

```python
plot(data_scaled, 'Detergents_Paper', 'Milk', 'Labels_DBSCAN2')
```

**Your comment here**

With a silhouette score of 0.5, DBSCAN seems to perform slightly better than K-Means, after optimizing the parameters of both models. This is expected, as DBSCAN is a mode more robust and able to cluster more complex distributions.
